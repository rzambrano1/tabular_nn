```python
import pandas as pd
import numpy as np
```

```python
def generate_categorical_table(n_rows=25, seed=42):
    np.random.seed(seed)
    
    # Define categorical levels
    sex_levels = ['Male', 'Female']
    region_levels = ['Northeast', 'Midwest', 'South', 'West']
    educ_levels = [
        "High School", 
        "Associate Degree", 
        "Bachelor's Degree", 
        "Master's Degree", 
        "Doctorate"
    ]
    
    # Generate data with weighted probabilities for realism
    data = {
        'sex': np.random.choice(sex_levels, size=n_rows, p=[0.49, 0.51]),
        'region': np.random.choice(region_levels, size=n_rows),
        'educ': np.random.choice(
            educ_levels, 
            size=n_rows, 
            p=[0.20, 0.15, 0.40, 0.20, 0.05]
        )
    }
    
    return pd.DataFrame(data)
```

```python
generate_categorical_table(5)
```

```python
generate_categorical_table(5).to_dict('records')
```
